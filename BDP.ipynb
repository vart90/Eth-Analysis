```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

### Q1 PART A

```python
with open('trans_output.txt', 'r') as f:
    lines = f.readlines()
data = ast.literal_eval(lines[0])
trans_df = pd.DataFrame(data,columns=['date','transactions'])
```

```python
a=[]

for i in trans_df['date'].values:
    l=i.split('/')
    a.append(l[1]+'/'+l[0])
    
    
trans_df['date']=a
```

```python
trans_df1 = trans_df.sort_values(by=['date'])
```

```python
# Set the figure size
plt.figure(figsize=(15, 6))
plt.bar(trans_df1['date'], trans_df1['transactions'], color='red')
plt.xlabel('Date')
plt.ylabel('Transactions')
plt.xticks(rotation=45)

plt.title('Number of Transactions by Date')
plt.show()
```

### Q4 Gas Guzzlers 

```python
with open('avg_gused.txt', 'r') as f:
    lines = f.readlines()
data = ast.literal_eval(lines[0])
gasuse_df = pd.DataFrame(data,columns=['date','gas'])
gasuse_df['gas'] = gasuse_df['gas'].astype(float)
```

```python
with open('avg_gprice.txt', 'r') as f:
    d = json.load(f)

d = [line.split(',') for line in d]    
gasprice_df = pd.DataFrame(d,columns = ['date','gas_price'])
gasprice_df['gas_price'] = gasprice_df['gas_price'].astype(float)
```

```python
c=[]

for i in gasuse_df['date'].values:
    l=i.split('/')
    c.append(l[1]+'/'+l[0])
    
    
gasuse_df['date']=c
gasuse_df1 = gasuse_df.sort_values(by=['date'])
```

```python
d=[]

for i in gasprice_df['date'].values:
    l=i.split('/')
    d.append(l[1]+'/'+l[0])
    
    
gasprice_df['date']=d
gasprice_df1 = gasprice_df.sort_values(by=['date'])
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 6))

ax[0].plot(gasuse_df1['date'], gasuse_df1['gas'], color='blue')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Avg Gas used')
ax[0].set_xticklabels(gasuse_df1['date'], rotation=90)
ax[0].set_title('Average Gas used by Date')

ax[1].plot(gasprice_df1['date'], gasprice_df1['gas_price'], color='blue')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Avg Gas Price')
ax[1].set_xticklabels(gasprice_df1['date'], rotation=90)
ax[1].set_title('Average Gas price by Date')

plt.show()
```

### Q4 Scam Analysis

#### Scams over time 

```python
import ast

with open('scam_time.txt', 'r') as f:
    d = ast.literal_eval(f.read())

scam_time_data = []
for entry in d:
    date, scam, val = entry[0][0], entry[0][1], entry[1]
    scam_time_data.append({'date': date, 'scam': scam, 'val': val})

scam_time_df = pd.DataFrame(scam_time_data, columns=['date', 'scam', 'val'])
display(scam_time_df)
```

```python
e=[]

for i in scam_time_df['date'].values:
    l=i.split('/')
    e.append(l[1]+'/'+l[0])
    

scam_time_df['date']=e
scam_time_df1 = scam_time_df.sort_values(by=['date'])

scam_time_df1
```

```python
scam_time_df1['year'] = pd.to_datetime(scam_time_df1['date'], format='%Y/%m').dt.year
scam_counts = scam_time_df1.groupby(['year', 'scam']).size().unstack(fill_value=0)

scam_counts.plot(kind='bar', stacked=True)
plt.xlabel('Year')
plt.ylabel('Number of Scams')
plt.title('Scam Counts By Year')
plt.show()
```

#### Most successful Scams 

```python
with open('top_scams.txt', 'r') as f:
    d = eval(f.read())

top_scams_df = pd.DataFrame([(entry[0][0], entry[0][1], entry[1]) for entry in d],
                            columns=['date', 'scam', 'val'])

display(top_scams_df)
```

```python
top_scams_df_new = top_scams_df.groupby(['scam']).size()
top_scams_df_new
```

```python
top_scams_df_new.plot(kind='bar')
plt.xlabel('Scam Type')
plt.ylabel('Number of Scams')
plt.title('Most successful scams')
plt.show()
```
